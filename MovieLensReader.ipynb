```python
def Ranker(dF,dM):
    ranks={}
    for k in dF.keys():
        if k in dM:
            if dM[k]>=dF[k]:
                ranks[k]=-dM[k]/dF[k]
            else:
                ranks[k]=dF[k]/dM[k]
        else:
            ranks[k]=dF[k]+len(ud)
    for k in dM.keys():
        if k not in dF:
            ranks[k]=-dM[k]-len(ud)
    return ranks
    
```

```python
def Selector(covranks, ranks, top=5):
    lF=[]
    lM=[]
    for k in ranks.keys():
        if ranks[k]>0:
            lF.append((covranks[k],k))
        else:    
            lM.append((covranks[k],k))
    lF=sorted(lF)
    lM=sorted(lM)
    lF.reverse()
    lM.reverse()
    return lF[:top],lM[:top]
```

```python
Selector(covs, ranks, top=5)
```

```python
def TriConApplicator(sel,tricontext):
    fMovies=[]
    mMovies=[]
    tcxt=[]
    for  t in sel[0]:
        fMovies.append(t[1])
    for  t in sel[1]:
        mMovies.append(t[1])   
    for t in tricontext:
        if t[1] in fMovies or t[1] in mMovies:
            tcxt.append(t)
    return tcxt
```
